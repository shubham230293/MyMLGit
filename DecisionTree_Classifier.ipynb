```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
```

```python
iris.data
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(iris.data, iris.target)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier, filled=True)
```

In the above plot we can also apply post pruning technique. At left node where X[2]<=1.65, we can do postpruning for no further split as value are [0,47,1] it will lead to overfitting only . and no further split is needed in this case. 

# Lets do some hyperparameter tuning and then apply plot the graph

```python
model = DecisionTreeClassifier(criterion='gini',
    splitter='best',
    max_depth=3,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features=None,
    random_state=None,
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    class_weight=None,
    ccp_alpha=0.0,
)
```

we have chosen criterion for split as gini impurity and max depth of tree as 3

```python
model.fit(iris.data, iris.target)
```

```python
plt.figure(figsize=(10,8))
tree.plot_tree(model, filled=True)
```

```python
#Lets apply entropy and then check the plot
model_ent = DecisionTreeClassifier(criterion='entropy',
    splitter='best',
    min_samples_split=2,
    min_samples_leaf=1)
```

```python
model_ent.fit(iris.data, iris.target)
```
